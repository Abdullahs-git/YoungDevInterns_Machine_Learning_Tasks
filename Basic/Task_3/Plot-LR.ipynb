```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)
X = np.linspace(1, 50, 50).reshape(-1, 1)
y = 3.5 * X + np.random.randn(50, 1) * 10
```

```python
model =LinearRegression()
model.fit(X,y)
```

```python
print("Intercept :" ,round(model.intercept_[0],2))
print("Slope (b1) :",round(model.coef_[0][0],2))
```

```python
y_pred=model.predict(X)
```

```python
mse=mean_squared_error(y,y_pred)
r2 =r2_score(y, y_pred)
print("\nModel Evaluation:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Task 3 - Regression Line and Predictions')
plt.xlabel('X (Input Feature)')
plt.ylabel('y (Target Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_new = np.array([[10], [25], [45]])
y_new_pred = model.predict(X_new)
```

```python
print("\nPredictions for New Data:")
for x_val, y_val in zip(X_new.flatten(), y_new_pred.flatten()):
    print(f"X = {x_val} → Predicted y = {y_val:.2f}")
```
