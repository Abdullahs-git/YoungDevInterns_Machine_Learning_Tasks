```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
x = np.random.rand(100) * 10
```

```python
y_positive = 2 * x + np.random.randn(100) * 3
```

```python
y_negative = -2 * x + np.random.randn(100) * 3 + 20
```

```python
y_none = np.random.rand(100) * 20
```

```python
df = pd.DataFrame({
    'x_feature': x,
    'y_positive': y_positive,
    'y_negative': y_negative,
    'y_no_correlation': y_none
})
```

```python
corr_positive = df['x_feature'].corr(df['y_positive'])
corr_negative = df['x_feature'].corr(df['y_negative'])
corr_none = df['x_feature'].corr(df['y_no_correlation'])
```

```python
print("Correlation Analysis")
print(f"Positive Correlation Coefficient (r): {corr_positive:.2f}")
print(f"Negative Correlation Coefficient (r): {corr_negative:.2f}")
print(f"No Correlation Coefficient (r):     {corr_none:.2f}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Visualizing Feature Correlations with Scatter Plots', fontsize=18)

#Plot 1
axes[0].scatter(df['x_feature'], df['y_positive'], alpha=0.7, edgecolors='k')
axes[0].set_title(f'Positive Correlation (r ≈ {corr_positive:.2f})', fontsize=14)
axes[0].set_xlabel('Feature X', fontsize=12)
axes[0].set_ylabel('Feature Y', fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.6)

#Plot 2
sns.regplot(data=df, x='x_feature', y='y_negative', ax=axes[1],
            scatter_kws={'alpha': 0.6, 'edgecolors':'k'}, line_kws={'color': 'red', 'linewidth': 2})
axes[1].set_title(f'Negative Correlation (r ≈ {corr_negative:.2f})', fontsize=14)
axes[1].set_xlabel('Feature X', fontsize=12)
axes[1].set_ylabel('Feature Y', fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.6)

#Plot 3
sns.regplot(data=df, x='x_feature', y='y_no_correlation', ax=axes[2],
            scatter_kws={'alpha': 0.6, 'edgecolors':'k'}, line_kws={'color': 'red', 'linewidth': 2})
axes[2].set_title(f'No Correlation (r ≈ {corr_none:.2f})', fontsize=14)
axes[2].set_xlabel('Feature X', fontsize=12)
axes[2].set_ylabel('Feature Y', fontsize=12)
axes[2].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
